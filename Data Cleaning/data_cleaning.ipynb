```python
import pandas as pd
```

```python
#read csv whose name is data.csv
df = pd.read_csv('data.csv')
print(df)
```

```python
#drop null values
df.dropna(inplace = True)
```

```python
#drop incorrect price types
for x in df.index:
    t=df.loc[x,'price']
    t=t.replace(',', '')
    df.loc[x,'price']=t
df['price'] = pd.to_numeric(df['price'])
df.dropna(subset=['price'], inplace = True)
```

```python
#drop incorrect rating types
df['rating'] = pd.to_numeric(df['rating'])
df.dropna(subset=['rating'], inplace = True)
```

```python
#drop incorrect reviewCount types
for x in df.index:
    t=df.loc[x,'reviewCount']
    t=t.replace(',', '')
    df.loc[x,'reviewCount']=t
df['reviewCount'] = pd.to_numeric(df['reviewCount'])
df.dropna(subset=['reviewCount'], inplace = True)
```

```python
#drop duplicates
df.drop_duplicates(inplace = True)
```

```python
print(df)
```

```python
df.to_csv("new_data.csv", header=True, index=True)
```
